```python
import matplotlib.pyplot as plt

def show_list(arr, title=""):
    fig, ax = plt.subplots(figsize=(20, 4))
    ax.bar(range(len(arr)), arr)
    ax.set_title(title)
    ax.set_xlabel('index')
    ax.set_ylabel('value')
    plt.show()
```

## Counting Sort
Bài toán: Counting Sort giải quyết bài toán sắp xếp dãy có ít giá trị riêng biệt.
Cách hoạt động: 

```python
number_list = [5, 9, 1, 10, 20, 4, 0, 0, 0, 20, 3]
show_list(number_list, title="number_list")
```

**Cách tiếp cận** của Counting Sort:
1. Duyệt qua dãy để đếm số lần xuất hiện của từng giá trị
2. Sử dụng các giá trị đếm được để đưa phần tử vào dãy mới theo thứ tự đã sắp xếp.

VD: Ta cần sắp xếp dãy [1,4,1,2,7,5,2] Chứa các giá trị từ 0 đến 7.
Bước 1: Tạo mảng *count* chứa số lần xuất hiện của mỗi phần tử:
               i:   0   1   2   3   4   5   6   7 
        count[i]:   0   2   2   0   1   1   0   1 

```python
def counting_sort(arr):
    # rà soát điều kiện
    if len(arr) <= 1:
        return 
    
    # Lấy ra giá trị max trong list
    max_val = max(arr) ## 7 => len(count) = 7
    
    ## Khởi tạo count list gồm max(arr) giá trị là 0
    count = [0] * (max_val+1)
    
    for el in arr:
        count[el] += 1 ## count[5] += 1
        
    ## Chuyển "count" về 'arr'
    index = 0
    for i in range(max_val+1): # Duyệt qua các giá trị có thể có từ 0 đến max_value
        for j in range(count[i]): # Số lần xuất hiện của giá trị i trong mảng gốc
            arr[index] = i # gán giá trị i vào vị trí tương ứng trong mảng sắp xếp
            index += 1 # tăng chỉ số của mảng đầu ra
    
```

```python
number_list_2 = number_list[:] # copy danh sách gốc

counting_sort(number_list_2)

print("Before sorting: {}".format(number_list))
print("After sorting: {}".format(number_list_2))
show_list(number_list_2, title="number_list_2")
```
